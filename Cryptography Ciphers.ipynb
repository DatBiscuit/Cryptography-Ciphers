```python
# function for finding the greatest divisor
def gcd(x,y):
    while y != 0:
        (x,y) = (y, x%y)
    return x

gcd(5,9)
```

```python
plainText = input("Input Plain Text here ")
shift = int(input("Shift number between 1-26 "))
shift = shift % 26

# function that uses a basic shift cipher
def shiftcipher(plainText, shift):
    # establish the alphabet and create the new alphabet
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    partOne = alphabet[:shift]
    partTwo = alphabet[shift:]
    newAlphabet = partTwo + partOne
    # initialize a variable to hold the cipher text
    cipherText = ""
    # iterate through the plain text and use the proper shift to create the cipher text
    for b in plainText:
        if b == " ":
            continue
        index = alphabet.find(b)
        cipherText += newAlphabet[index]
    return cipherText

shiftcipher(plainText, shift)
```

```python
cipherText = input("Input Cipher Text here ")

# brute force for a shift cipher / caesars cipher
def decryptshiftcipher(cipherText):
    #establish the alphabet
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    # a is used as our shift value, we increase this every time in the while loop
    a = 0
    while a < 26:
        attempt = ""
        partOne = alphabet[:a]
        partTwo = alphabet[a:]
        newAlphabet = partTwo + partOne
        # after creating our new alphabet with our shift value we iterate through the
        # plain text and print out the attempt to see if we get anything closely
        # relating to english, instead of a bunch of gibberish
        for b in cipherText:
            index = newAlphabet.find(b)
            attempt += alphabet[index]
        
        a += 1
        print(attempt)
        
decryptshiftcipher(cipherText)
```

```python
# a function that implements an affineCipher
# y -> x*alpha + beta (mod26)
def affineCipher():
    plainText = input("Enter your text to encrypt here: ")
    alpha = int(input("Enter an alpha: "))
    beta = int(input("Enter a beta: "))
    # want to ensure that the alpha is relatively prime to 26
    # otherwise ambiguity can become present
    while gcd(alpha, 26) != 1:
        print("Alpha must be relatively prime to 26 try again")
        alpha = int(input("Enter an alpha: "))
    
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    
    # iterate through the plain text and shift it according to the affine function
    cipherText = ""
    for y in plainText:
        if y == " ":
            continue
        x = alphabet.find(y)
        index = (alpha * x + beta) % 26
        cipherText += alphabet[index]
    return cipherText
        
affineCipher()
```

```python
# a brute force function to decrypt an Affine Cipher
# since there is only 312 possibilities brute force is alright to use instead of
# frequency analysis
def decryptAffineCipher(cipherText):
    # establish our alphabet
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    # we only need to check for an alpha that is relatively prime to 26
    # x is our alpha
    for x in range(25):
        if gcd(x,26) != 1:
            continue
        print("alpha = " + str(x))
        # once we find an alpha we go through all 26 options for beta
        for y in range(25):
            attempt = ""
            print("beta = " + str(y))
            for char in range(len(cipherText)):
                # function to reverse an affine cipher:
                # (1/alpha)(y - beta) % 26; where 1/alpha is the modular inverse
                index = ((mod_inverse(x,26))*((alphabet.find(cipherText[char])) - y)) % 26
                attempt += alphabet[index]
            print(attempt)
                
decryptAffineCipher("cvvwpm")
```

```python
# a function to implement the vigenere cipher
# this cipher uses a "key" or word to help with encryption rather than a single number
# also implements a shift but uses a word so depending on which letter of the key that
# we are looking at, that is the one that will be our shift 
def vigenereCipher():
    plainText = input("Enter text for encryption here: ")
    key = input("Enter the key(word) for encryption: ")
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    list = []
    cipherText = ""
    # create a list to store the values for our key
    for i in key:
        index = alphabet.find(i)
        list.append(index)
    # this variable is used in case our key is shorter than our plaintext
    # the key will keep being repeated until it covers all of the plaintext
    spot = 0
    # this loop is used for turning our plain text into cipher text
    for j in plainText:
        # if our spot variable is greater than the length of key we have to put it back
        # to zero
        if spot >= len(key):
            spot = 0
        index = alphabet.find(j)
        # grab the value of the key add it to the index then mod it by 26
        newIndex = (list[spot] + index) % 26
        cipherText += alphabet[newIndex]
        spot += 1
    return cipherText
        

vigenereCipher()
```

```python
# need to import the random library in order to create our matrices
# importing numpy for matrix multiplication 
import random as r
import numpy as np

# this implements a block cipher which uses matrix of n by nth dimension and multiply it
# by the plain text
def blockCipher():
    # gathers the n by n matrix value
    intn = int(input("Choose an integer so they key is an n x n matrix M whose entries"+
                     "are integers mod 26"))
    plainText = input("Enter plain text here: ")
    matrix = []
    # iterates through to randomize the matrix
    for i in range(intn):
        temp = []
        for j in range(intn):
            temp.append(r.randint(1,25))
        matrix.append(temp)
    print(matrix)
    
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    plain = [] # transform the plain text into a number matrix of 1 x number of characters
    for p in plainText:
        plain.append(alphabet.find(p))
        
    # need to mod it by 26 so we stay in the range of the alphabet
    cipher = (np.matmul(plain, matrix)) % 26
    cipherText = ""
    for c in cipher:
        cipherText += alphabet[int(c)]
        
        
    return cipherText
        
blockCipher()
    
    
```
